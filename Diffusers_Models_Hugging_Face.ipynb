## Generating faces using diffusion models

```python
from diffusers import DDPMPipeline # denoising diffusion probabilistic model
```

```python
#youll need to run with gpu for cuda to work
ddpm = DDPMPipeline.from_pretrained("google/ddpm-celebahq-256").to('cuda')
```

```python
image = ddpm(num_inference_steps=30).images[0]
image
```

```python
from diffusers import DDPMScheduler, UNet1DModel
```

```python
scheduler = DDPMScheduler.from_pretrained("google/ddpm-celebahq-256")
```

```python
scheduler
```

```python
model = UNet2DModel.from_pretrained("google/ddpm-celebahq-256").to("cuda")
```

```python
model
```

```python
scheduler.set_timesteps(50)
```

```python
sample_size = model.config.sample_size
```

```python
sample_size
```

```python
import torch
```

```python
noise = torch.randn((1, 3, sample_size, sample_size),device ="cuda")
```

```python
noise
```

```python
input = noise
input.squeeze()
```

```python
for t in scheduler.timesteps:
  with torch.inference_mode():
    noisy_residual = model(input,t).sample

    previous_noisy_sample = scheduler.step(noisy_residual, t, input).prev_sample

    input = previous_noisy_sample
```

```python
from PIL import Image
import numpy as np
```

```python
image = (input / 2 + 0.5).clamp(0,1).squeeze()
```

```python
image = (image.permute(1,2,0)*255).round().to(torch.uint8).cpu().numpy()
```

```python
image
```

## Generating images using a prompt

```python
import torch
from diffusers import DiffusionPipeline

pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.bfloat16, device_map="cuda")

```

```python
prompt = "Astronaut in a jungle, cold color palette, muted colors, detailed, 8k"
image = pipe(prompt).images[0]
```
