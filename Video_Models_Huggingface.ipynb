## Generating video with diffusion

```python
from diffusers import StableVideoDiffusionPipeline
```

```python
pipe = StableVideoDiffusionPipeline.from_pretrained("stabilityai/stable-video-diffusion-img2vid-xt", torch_dtype=torch.float16,variant = "fp16", use_safetensors=True)
```

```python
pipe
```

```python
image_path = "images.jpg"
```

```python
from diffusers.utils import load_image, export_to_video
```

```python
image = load_image(image_path)
image
```

```python
pipe.enable_model_cpu_offload()
```

```python
generator = torch.manual_seed(42)
```

```python
generator
```

```python
frames = pipe(image, decode_chunk_size=8,generator=generator).frames[0]
```

```python
frames
```

```python
export_to_video(frames,"cats.mp4",fps=7)
```

##Generating Video with another diffusion model(using image and a prompt)

```python
import torch
torch.cuda.empty_cache()
```

```python
from diffusers import I2VGenXLPipeline
```

```python
from diffusers.utils import load_image,export_to_video
```

```python
repo_id = "ali-vilab/i2vgen-xl"
```

```python
pipeline = I2VGenXLPipeline.from_pretrained(repo_id,
                                             torch_dtype=torch.float16,
                                             variant="fp16")
```

```python
pipeline.enable_model_cpu_offload()
```

```python
img = load_image(image_path)
```

```python
prompt = "animate the cat to move its head a bit"
```

```python
generator = torch.manual_seed(42)
```

```python
frames = pipeline(prompt = prompt,
                  image = img,
                  num_frames = 16,
                  generator=generator).frames[0]
```

```python
frames
```

```python
export_to_video(frames,"vid.mp4",fps = 4)
```
