```python
def multiply(x,y):
  return x*y
```

there are 2 types of events:
1. click events
2. chain events



## Change event

```python
with gr.Blocks() as demo:
  with gr.Row():
    x_slider = gr.Slider()
    y_slider = gr.Slider()

  with gr.Row():
    result = gr.Text()

  y_slider.change(fn=multiply,
                  inputs = [x_slider,y_slider],
                  outputs=result)

  demo.launch()
```

```python
with gr.Blocks() as demo:
  with gr.Row():
    x_slider = gr.Slider(label='X')
    y_slider = gr.Slider(label='Y')

  with gr.Row():
    result = gr.Text()

  x_slider.change(fn=multiply,
                  inputs=[x_slider,y_slider],
                  outputs=result)
  y_slider.change(fn=multiply,
                  inputs=[x_slider,y_slider],
                  outputs=result)

  demo.launch()
```

## click event

```python
with gr.Blocks() as demo:
  with gr.Row():
    x_slider = gr.Slider(label='X')
    y_slider = gr.Slider(label='Y')

  with gr.Row():
    result = gr.Text()

  with gr.Row():
    button = gr.Button(value='multiply')

  button.click(fn=multiply,
               inputs=[x_slider,y_slider],
               outputs = result)

  demo.launch()
```

## returning different outputs

```python
def make_greyscale(image_path):

  image = Image.open(image_path)

  image_grey = image.convert('L')

  return image_grey, "Image converted"
```

```python
img,log = make_greyscale("images.jpg")
```

```python
with gr.Blocks() as demo:
  with gr.Row():
    img = gr.Image(type='filepath')
    output = gr.Image()

  with gr.Row():
    log = gr.Text()
    submit = gr.Button(value = "convert to grey")

  submit.click(fn=make_greyscale,
               inputs = img,
               outputs=[output,log])

  demo.launch()

```

## Gradion integration with ML

```python
from transformers import AutoImageProcessor, AutoModelForImageClassification

processor = AutoImageProcessor.from_pretrained("microsoft/resnet-18")
model = AutoModelForImageClassification.from_pretrained("microsoft/resnet-18")
```

```python
model
```

```python
def classifying_img(image):
  image = processor(image, return_tensor="pt")["pixel_values"]

  logits = model(image).logits

  predicted_id = logits.argmax(-1).item()

  return model.config.id2label[predicted_id]
```

```python
iface = gr.Interface(rn=classifying_img,
                     imputs=gr.Image(),
                     outputs=gr.Label(),
                     description="classifies a image")
iface.launch()
```

## Gradient sentiment analysis

```python
from transformers import pipeline
```

```python
sentiment_analysis = pipeline("sentiment-analysis")
```

```python
def sentiment(text):
  result = sentiment_analysis(text)
  return result[0]['label'], result[0]['score']
```

```python
iface = gr.Interface(fn=sentiment_analysis,
                     inputs=gr.Textbox(lines=2),
                     outputs=[gr.Text(label="sentiment"),gr.Text(label="score")])
iface.launch()
```

## Gradial errors

```python
def validate(num):
  return num%2==0
```

```python
def add_even(a,b):
  if not validate(a) or not validate(b):
    raise gr.Error("inputs must be even")
  return a+b
```

```python
iface = gr.Interface(fn=add_even,
                     inputs=[gr.Number(),gr.Number()],
                     outputs=gr.Number())
iface.launch()
```
