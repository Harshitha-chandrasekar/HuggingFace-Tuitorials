```python
import torch
import torch.nn.functional as F
```

```python
# top k highest possibilities from the next values
scentence = "i learn machine learning to destroy human"
tokens = tokenizer(scentence, return_tensors='pt').input_ids
outputs = gpt2(tokens).logits[0, -1]
final_logits = torch.topk(outputs,10)
for index in final_logits.indices:
    print(tokenizer.decode(index))
```

```python
def greedy_decode(logits):
  #return token index with max probability
  return torch.argmax(logits,dim=-1)

greedy_decode(outputs)
```

```python
def top_k(logits, k = 10):
  # keeps only top k logits normalize them into probabilities
  # then sample 1 token from the filtered distribution
  values, indices = torch.topk(logits, k)
  probs = F.softmax(values, dim = -1)
  samples = torch.multinomial(probs, 1)
  return indices[samples]

tokenizer.decode(top_k(outputs))
```

```python
def top_p(logits, p = 0.9):
  # sort tokens by prob. keep smallest no whose cumulative prob value exceeds p,
  # then sameple 1 token
  sorted_logits, sorted_indices = torch.sort(logits, descending=True)
  sorted_probs = F.softmax(sorted_logits, dim = -1)
  cumulative_probs = sorted_probs.cumsum(dim = -1)

  # mask tokens not being considered
  mask = cumulative_probs > p
  sorted_logits[mask] = float('-inf')

  # sample from filtered logits
  filtered_probs = F.softmax(sorted_logits, dim = -1)
  sampled = torch.multinomial(filtered_probs, 1)

  # return token index in original vocabulary
  return sorted_indices[sampled]

tokenizer.decode(top_p(outputs))
```

```python
def temperature_sampling(logits, temperature =0.8):
  # scales logits by temp before sampling
  # lower temp  -> sharper more deterministic distributuin
  # higher temp -> flatter temp distribution

  scaled = logits / temperature
  probs = F.softmax(scaled, dim = -1)
  return torch.multinomial(probs, 1)

tokenizer.decode(temperature_sampling(outputs))
```

```python
def random_sampling(logits):
  # samples directly from softmax dist without filtering

  probs = F.softmax(logits, dim = -1)
  return torch.multinomial(probs, 1)

tokenizer.decode(random_sampling(outputs))
```

```python
scentence = "I learn Maching Learning to enhance"
tokens_ids = tokenizer(scentence, return_tensors='pt').input_ids
outputs = gpt2(tokens_ids).logits[0, -1]
outputs = torch.softmax(outputs, dim = -1)

top10 = torch.topk(outputs, k = 10)

for index, values in zip(top10.indices, top10.values):
  print(f"{tokenizer.decode(index)} --  {values:.1%}")
```
